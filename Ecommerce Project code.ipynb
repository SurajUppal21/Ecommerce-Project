## Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.
**Seting a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column.**

```python
customers.columns
X = customers[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
X.head()
```

```python
y = customers["Yearly Amount Spent"]
y.head()
```

**Using model_selection.train_test_split from sklearn to split the data into training and testing sets. Seting test_size=0.3 and random_state=101**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

## Training the Model

Now its time to train our model on our training data!

**Importing LinearRegression from sklearn.linear_model**

```python
from sklearn.linear_model import LinearRegression
```

**Create an instance of a LinearRegression() model named lm.**

```python
lm = LinearRegression()
```

**Train/fit lm on the training data**

```python
lm.fit(X_train,y_train)
```

**Print out the coefficients of the model**

```python
print("Coefficients: \n",lm.coef_)
pd.DataFrame(lm.coef_,index = X.columns,columns=["Coeff"])
```

## Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

**Use lm.predict() to predict off the X_test set of the data.**

```python
predictedy = lm.predict(X_test);
```

**Create a scatterplot of the real test values versus the predicted values.**

```python
plt.xlabel("Y Test(True Values)")
plt.ylabel("Predicted Y")
z = sns.scatterplot(x=y_test,y=predictedy)
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

**Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. Refer to the lecture or to Wikipedia for the formulas**

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(predictedy,y_test))
print("MSE:",metrics.mean_squared_error(predictedy,y_test))
print("RMSE:",np.sqrt(metrics.mean_squared_error(predictedy,y_test)))
```

```python
metrics.explained_variance_score(y_test,predictedy)#r^2 is the measure of how much variance our model
#explains this means our model here explain about 98.9% of variance
#we dont need to worry about the residuals because our model has a very good fit
```

## Residuals

We should have gotten a very good model with a good fit. Let's quickly explore the residuals to make sure everything was okay with our data. 

**Plot a histogram of the residuals and make sure it looks normally distributed. Use either seaborn distplot, or just plt.hist().**

```python
sns.distplot(y_test-predictedy,kde=True,bins=50,color="black",kde_kws={"alpha":0.6,"lw":2})
```
